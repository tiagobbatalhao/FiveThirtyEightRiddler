```python
import pylab as py
import networkx
```

```python
# Uploading data
fl = open('scores.htm')
games = []
for line in fl.readlines():
    if line[0:3]=='201':
        this = line.split(' ')
        date = line[:10]
#         teamA = line[10:36].strip(' @')
        
        digits = [x for x in range(len(line)) if line[x] in '0123456789' and x>=10]
        digits_groups = [[digits[0]]]
        previous = digits[0]
        for dig in digits[1:]:
            if dig-previous==1:
                digits_groups[-1].append(dig)
            else:
                digits_groups.append([dig])
            previous = dig + 0
    
        digitsA = digits_groups[0]
        digitsB = digits_groups[1]
        teamA = line[10:digitsA[0]].strip(' @')
        teamB = line[digitsA[-1]+1:digitsB[0]].strip(' @')
        scoreA = int(''.join([line[x] for x in digitsA]))
        scoreB = int(''.join([line[x] for x in digitsB]))
        this = (date,teamA,teamB,scoreA,scoreB)
        games.append(this)
```

```python
teams = set([x[1] for x in games]+[x[2] for x in games])
graph = networkx.MultiDiGraph()
for team in teams:
    graph.add_node(team)
```

```python
for game in games:
    graph.add_edge(game[1],game[2], date=game[0])
```

```python
print(networkx.info(graph))
```

```python
champions = [ x for x in teams if networkx.has_path(graph, x, 'Villanova')]
```

```python
for team in champions:
    print(team, networkx.shortest_path(graph, team, 'Villanova'))
    
```
